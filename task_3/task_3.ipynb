```python
import pandas as pd
```

```python
# Load supplier data
supplier_df1 = pd.read_excel(r"C:\Users\SwadeshRanjanPanda\Desktop\resources\resources\task_3\supplier_data1.xlsx")
supplier_df2 = pd.read_excel(r"C:\Users\SwadeshRanjanPanda\Desktop\resources\resources\task_3\supplier_data2.xlsx")
supplier_df = pd.concat([supplier_df1, supplier_df2],  axis=1)
supplier_df.head()
pd.set_option('display.max_columns', None)
```

```python
#Adding both the column Quantity to Total_Quantity
supplier_df['Total Quantity'] = supplier_df1['Quantity'] + supplier_df2['Quantity']
supplier_df.head()
```

```python
# Load buyer preferences
buyer_df = pd.read_excel(r"C:\Users\SwadeshRanjanPanda\Desktop\resources\resources\task_3\buyer_preferences.xlsx")
buyer_df.head()
```

```python
# Perform matching based on criteria
recommendations = buyer_df.merge(supplier_df,
                                 left_on=["Preferred Grade", "Preferred Finish"],
                                 right_on=["Grade", "Finish"],
                                 how="inner")
recommendations
```

```python
# Apply additional conditions to filter based on numerical columns
# filtered_df = recommendations[
#     (recommendations["Thickness (mm)"] >= recommendations["Preferred Thickness (mm)"]) &
#     (recommendations["Width (mm)"] >= recommendations["Preferred Width (mm)"]) &
#     (recommendations["Gross weight (kg)"] >= recommendations["Max Weight (kg)"]) &
#     (recommendations["Total Quantity"] >= recommendations["Min Quantity"])
# ]
# Apply additional conditions to filter based on numerical columns
filtered_df = recommendations[
    (recommendations["Thickness (mm)"] <= recommendations["Preferred Thickness (mm)"]) &
    (recommendations["Gross weight (kg)"] <= recommendations["Max Weight (kg)"]) &
    (recommendations["Total Quantity"] >= recommendations["Min Quantity"])
]
```

```python
filtered_df
```

```python
filtered_df.rename(columns={"Thickness (mm)": "Thickness_mm", "Width (mm)": "Width_mm", "Gross weight (kg)":"Gross_weight_kg"}, inplace=True)
```

```python
# Keep only relevant columns
recommendations_final = filtered_df[["Buyer ID", "Material", "Grade", "Finish", "Thickness_mm", "Width_mm", "Total Quantity", "Gross_weight_kg"]]
recommendations_final
```
