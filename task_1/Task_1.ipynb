```python
# Step 1: Find rows where MATERIAL_QUALITY_NORM is NaN
missing_quality_norms = supplier_data_2[supplier_data_2["MATERIAL_QUALITY_NORM"].isna()][["MATERIAL_NAME", "MATERIAL_QUALITY_NORM"]]

# Step 2: Find known MATERIAL_QUALITY_NORM values for the same MATERIAL_NAME
previous_values = supplier_data_2.dropna(subset=["MATERIAL_QUALITY_NORM"]).drop_duplicates(subset=["MATERIAL_NAME"])

# Step 3: Merge missing rows with known values
supplier_data_2 = supplier_data_2.merge(previous_values[["MATERIAL_NAME", "MATERIAL_QUALITY_NORM"]], on="MATERIAL_NAME", how="left", suffixes=("_current", "_previous"))

# Step 4: Fill NaN values in MATERIAL_QUALITY_NORM with previous known values
supplier_data_2["MATERIAL_QUALITY_NORM_current"] = supplier_data_2["MATERIAL_QUALITY_NORM_current"].fillna(supplier_data_2["MATERIAL_QUALITY_NORM_previous"])

# Step 5: Drop the extra column and rename back
supplier_data_2.drop(columns=["MATERIAL_QUALITY_NORM_previous"], inplace=True)
supplier_data_2.rename(columns={"MATERIAL_QUALITY_NORM_current": "MATERIAL_QUALITY_NORM"}, inplace=True)

# Display the final dataset
print(supplier_data_2[supplier_data_2["MATERIAL_QUALITY_NORM"].isna()])  # Check if any NaNs are still left
```
